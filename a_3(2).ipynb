```
import pandas as pd
import matplotlib.pyplot as plt
d=pd.read_csv('/content/mobile_cleaned-1549119762886.csv')
```

```
a=list(d.columns)
for i in range(len(d.columns)):
  for n in a:
    mean=d[n].mean()
    d[n].fillna(mean,inplace=True)
y=list(d.price)
for n in a:
  x=list(d[n])
  plt.ylabel("price")
  plt.xlabel(n)
  plt.plot(x,y,label=n)
  plt.legend()
  plt.show()
  
```
